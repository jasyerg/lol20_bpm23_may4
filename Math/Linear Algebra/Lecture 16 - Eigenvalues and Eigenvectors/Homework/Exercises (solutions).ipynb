### Problem 1:

Find the eignevalues and eignevectors for 

\begin{equation}
\pmb A = 
\begin{bmatrix}
    4 & 2\\
    3 & -1
\end{bmatrix}
\end{equation}


### SOLUTION

__Step 1:__ Rearrange the equation $\pmb A \pmb x = \lambda \pmb x$:
<center> $\pmb A \pmb x - \lambda \pmb x = 0$ </center> 

__Step 2:__ Factor out the x:
<center> $(\pmb A - \lambda \pmb I)\pmb x = 0$ </center> 

__Step 3:__ Find $det(\pmb A - \lambda \pmb I)$ in terms of the elements of the matrix 
\begin{equation}
\pmb A - \lambda \pmb I = 
\begin{bmatrix}
    4 & 2\\
    3 & -1
\end{bmatrix}
-
\begin{bmatrix}
    \lambda & 0\\
    0 & \lambda
\end{bmatrix}
=
\begin{bmatrix}
    4 - \lambda \pmb I & 2\\
    3 & -1 - \lambda \pmb I
\end{bmatrix}
\end{equation}

\begin{equation}
det(\pmb A - \lambda \pmb I)
= 
(4 - \lambda)(-1 - \lambda \pmb I) - (2)(3) = \lambda^2 - 3\lambda - 10 
\end{equation}

__Step 4:__ Set $\lambda^2 - 3\lambda - 10 = 0$ and solve for the $\lambda$(s) which are the Eigenvalues of the matrix $\pmb A$ $\Rightarrow$ $\lambda_1 = 5$ and $\lambda_2 = -2$.

__Step 5:__ Plug $\lambda$(s) back into $\pmb A - \lambda \pmb I$ to create 2 new matrices:
\begin{equation}
\pmb A - 5\pmb I = 
\begin{bmatrix}
    -1 & 2\\
    3 & -6
\end{bmatrix}
\end{equation}

\begin{equation}
\pmb A + 2\pmb I = 
\begin{bmatrix}
    6 & 2\\
    3 & 1
\end{bmatrix}
\end{equation}

__Step 6:__ Solve the system of equations in the 2 new matrices using Gaussian Elimination and obtaining RREF

\begin{equation}
\begin{array}{cc|c}  
    -1 & 2 & 0\\
    -1 & 2 & 0\\
\end{array}
\Rightarrow x_1 = 
\begin{bmatrix}
    2 \\
    1
\end{bmatrix}
\end{equation}  

\begin{equation}
\begin{array}{cc|c}  
    3 & 1 & 0\\
    3 & 1 & 0\\
\end{array}
\Rightarrow x_2 = 
\begin{bmatrix}
    -1 \\
    3
\end{bmatrix}
\end{equation}
