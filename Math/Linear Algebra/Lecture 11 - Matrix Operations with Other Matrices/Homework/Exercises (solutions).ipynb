### Problem 1

Given the following matrices:

$
\pmb A = 
\begin{bmatrix}
    -1 & 3 & 2\\
    0 & -2 & 3\\
\end{bmatrix}
$

$
\pmb B = 
\begin{bmatrix}
    1 & 3 & -1\\
    3 & -2 & 0\\
\end{bmatrix}
$

$
\pmb C = 
\begin{bmatrix}
    1 & 2 \\
    4 & 3\\
\end{bmatrix}
$

$
\pmb D = 
\begin{bmatrix}
    3 & 1 \\
    -1 & 2\\
\end{bmatrix}
$

Perform the following manual calculations:

a) $\pmb A + \pmb B$

b) $\pmb C + \pmb D$

c) $\pmb A \pmb B^T$

d) $\pmb C \pmb D$

e) $\pmb C \pmb A$

### SOLUTIONS

a)
$
\pmb A + \pmb B = 
\begin{bmatrix}
    0 & 6 & 1\\
    3 & -4 & 3\\
\end{bmatrix}
$

b)
$
\pmb C + \pmb D = 
\begin{bmatrix}
    4 & 3 \\
    3 & 5 \\
\end{bmatrix}
$

c) $\pmb A \pmb B^T =
\begin{bmatrix}
    -1+9-2 & 0-6+6 \\
    0-6-3 & 0+4+9 \\
\end{bmatrix} = 
\begin{bmatrix}
    6 & 0 \\
    -9 & 13 \\
\end{bmatrix}
$

d) $\pmb C \pmb D =
\begin{bmatrix}
    3 -2 & 1+4 \\
    12-3 & 4+6 \\
\end{bmatrix} = 
\begin{bmatrix}
    1 & 5 \\
    9 & 10 \\
\end{bmatrix}
$

e) $\pmb C \pmb A =
\begin{bmatrix}
    -1+0 & 3-4 & 2+6 \\
    -4+0 & 12-6 & 8+9 \\
\end{bmatrix} = 
\begin{bmatrix}
    -1 & -1 & 8 \\
    -4 & 6 & 17 \\
\end{bmatrix}
$
